**Cleaning of the data**

```python
data.replace('?', pd.NA, inplace=True) #replace all occurrences of the question mark character '?' in the DataFrame
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['city-mpg'] = pd.to_numeric(data['city-mpg'], errors='coerce')
data['highway-mpg'] = pd.to_numeric(data['highway-mpg'], errors='coerce')
data['peak-rpm'] = pd.to_numeric(data['peak-rpm'],errors='coerce') #used to convert the values in the 'peak-rpm' column of the DataFrame
data['curb-weight'].fillna(data['curb-weight'].mean(), inplace=True) #fill missing values in the 'curb-weight' column
data['make'].fillna('Unknown', inplace=True)
data['drive-wheels'].fillna('Unknown', inplace=True)
doors_mapping = {'two': 2, 'four': 4}
data['num-of-doors'] = data['num-of-doors'].map(doors_mapping)

```

Let's start with visualisation

**Vehicle make frequency diagram according to Makers
**

```python
make_count = data['make'].value_counts()
plt.figure(figsize=(10, 6))
make_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frequency of Vehicle Makes')
plt.xlabel('Vehicle Make')
plt.ylabel('Count')
plt.xticks(rotation=45)

for i, count in enumerate(make_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

**Insurance risk ratings Histogram**

```python
data.symboling.hist(bins=6,color='green');
plt.title("Insurance risk ratings of vehicles")
plt.ylabel('Number of vehicles')
plt.xlabel('Risk rating');
```

**Normalized losses histogram**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['normalized-losses'].dropna(), bins=30, kde=True)
plt.title('Distribution of Normalized Losses')
plt.xlabel('Normalized Losses')
plt.ylabel('Frequency')
plt.show()
```

**Fuel type bar chart**

```python
fuel_type_count = data['fuel-type'].value_counts()
plt.figure(figsize=(10, 6))
fuel_type_count.plot(kind='bar', color=['blue', 'green', 'red', 'purple'])
plt.title('Distribution of Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
for i, count in enumerate(fuel_type_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

**Fuel type pie diagram**

```python
fuel_counts = data['fuel-type'].value_counts()
plt.figure(figsize=(8, 8))
fuel_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['green','blue'])
plt.title('Distribution of Fuel Types')
plt.ylabel('')  # Hide the y-label
plt.show()
```

**Horse power histogram**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='horsepower', y='num-of-cylinders')
plt.title('Horsepower Distribution by Number of Cylinders')
plt.xlabel('Horsepower')
plt.ylabel('Number of Cylinders')
plt.show()
```

**Curb weight histogram**

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='make', y='curb-weight')
plt.title('Curb Weight Distribution by Vehicle Make')
plt.xlabel('Vehicle Make')
plt.ylabel('Curb Weight')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

**Drive wheels bar chart**

```python
drive_wheels_count = data['drive-wheels'].value_counts()
plt.figure(figsize=(10, 6))
drive_wheels_count.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Distribution of Drive Wheels')
plt.xlabel('Drive Wheels')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, count in enumerate(drive_wheels_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

**Number of doors bar chart**

```python
mean_price_doors = data.groupby('num-of-doors')['price'].mean()
ax = mean_price_doors.plot(kind='bar', color='peru')
plt.title("Number of Doors vs Price")
plt.ylabel('Price')
plt.xlabel('Number of Doors')
for i, v in enumerate(mean_price_doors):
    plt.text(i, v + 1000, f'${v:,.2f}', ha='center', va='bottom')
plt.show()
```

**Scatter plot of City and Highway MPG, Curb weight based on Make of the car**

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='curb-weight', y='highway-mpg',  hue='make', data=data, palette='viridis', alpha=0.7, sizes=(20, 200))
plt.title('Scatter Plot of Highway MPG based on curb-weight')
plt.xlabel('curb-weight')
plt.ylabel('Highway MPG')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='curb-weight', y='city-mpg', hue='make', data=data, palette='viridis', alpha=0.7)
plt.title('Scatter Plot of City MPG vs Curb Weight')
plt.xlabel('Curb Weight')
plt.ylabel('City MPG')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Make')
plt.show()
```

Based on the scatter plot and the provided data set, it is clear that for both city and highway mileage of the automobile is inversely proportional to the curb weight.

Heavier the Automobile less is the mileage for both City and Highway
